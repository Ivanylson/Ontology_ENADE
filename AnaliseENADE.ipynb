 Código tem como objetivo modificar uma ontologia RDF existente relacionada aos dados do ENADE, adicionando instâncias e informações de proveniência utilizando o padrão PROV-O (Ontology for Provenance) da W3C.

```python
from rdflib import Graph, Namespace, Literal, RDF, RDFS, URIRef
import pandas as pd
from rdflib.namespace import XSD

# 1. Carregar a ontologia existente
ontology_path = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/enade.owl'
g = Graph()
g.parse(ontology_path, format='xml')

# 2. Definir namespaces
EX = Namespace("http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022#")
PROV = Namespace("http://www.w3.org/ns/prov#")
g.bind("ex", EX)
g.bind("prov", PROV)

# 3. Adicionar novas classes e propriedades PROV-O
g.add((PROV['Entity'], RDF.type, RDFS.Class))
g.add((PROV['Activity'], RDF.type, RDFS.Class))
g.add((PROV['Agent'], RDF.type, RDFS.Class))
g.add((PROV['Usage'], RDF.type, RDF.Property))
g.add((PROV['WasGeneratedBy'], RDF.type, RDF.Property))
g.add((PROV['WasAssociatedWith'], RDF.type, RDF.Property))
g.add((PROV['WasDerivedFrom'], RDF.type, RDF.Property))

# Adicionar as propriedades PROV-O às classes existentes
g.add((EX['Prova'], RDFS.subClassOf, PROV['Entity']))
g.add((EX['Resposta'], RDFS.subClassOf, PROV['Entity']))
g.add((EX['Item'], RDFS.subClassOf, PROV['Entity']))

# 4. Ler os dados do arquivo CSV
file_path_arq3 = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/microdados2022_arq3.txt'
df = pd.read_csv(file_path_arq3, sep=';', encoding='latin1')

# 5. Adicionar instâncias e proveniência à ontologia
for _, row in df.iterrows():
    # Criar URIs para as instâncias
    prova_uri = URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Prova/{row['CO_CURSO']}_{row['NU_ANO']}")
    resposta_uri = URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Resposta/{row['CO_CURSO']}_{row['NU_ANO']}_{row['NU_ITEM_OFG']}")
    item_uri = URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Item/{row['NU_ITEM_OFG']}")
    entity_uri = URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Entity/{row['CO_CURSO']}_{row['NU_ANO']}_{row['NU_ITEM_OFG']}")

    # Adicionar instâncias
    g.add((prova_uri, RDF.type, EX['Prova']))
    g.add((resposta_uri, RDF.type, EX['Resposta']))
    g.add((item_uri, RDF.type, EX['Item']))
    g.add((entity_uri, RDF.type, PROV['Entity']))

    # Adicionar propriedades
    g.add((prova_uri, EX['temItem'], item_uri))
    g.add((resposta_uri, EX['temNota'], Literal(row['NT_GER'], datatype=XSD.decimal)))
    g.add((resposta_uri, RDFS.comment, Literal(row['DS_VT_GAB_OFG_FIN'], datatype=XSD.string)))

    # Adicionar dados de proveniência
    g.add((entity_uri, PROV['wasGeneratedBy'], prova_uri))
    g.add((prova_uri, PROV['wasAssociatedWith'], URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Agent/{row['CO_CURSO']}")))
    g.add((entity_uri, PROV['wasDerivedFrom'], URIRef(f"http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022/Source/{row['CO_CURSO']}_{row['NU_ANO']}")))

# 6. Salvar a ontologia modificada
output_path = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/enade2022_updated_with_prov.owl'
g.serialize(destination=output_path, format='xml')

print(f"Ontologia modificada com PROV-O salva em {output_path}")
```

O código carrega uma ontologia atualizada em formato OWL, define namespaces, executa consultas SPARQL para verificar entidades e suas propriedades relacionadas ao padrão PROV-O, e exibe os resultados dessas consultas, incluindo informações sobre como as entidades foram geradas, associadas e derivadas.


```python
from rdflib import Graph, Namespace, URIRef
from rdflib.plugins.sparql import prepareQuery

# Carregar a ontologia atualizada
ontology_path = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/enade2022_updated_with_prov.owl'
g = Graph()
g.parse(ontology_path, format='xml')

# Definir namespaces
EX = Namespace("http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022#")
PROV = Namespace("http://www.w3.org/ns/prov#")
g.bind("ex", EX)
g.bind("prov", PROV)

# Consulta SPARQL para verificar entidades e suas propriedades PROV-O
query = """
PREFIX ex: <http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022#>
PREFIX prov: <http://www.w3.org/ns/prov#>

SELECT ?entity ?property ?value
WHERE {
    ?entity a prov:Entity .
    ?entity ?property ?value .
}
"""

# Executar a consulta
results = g.query(query)

# Exibir resultados
for row in results:
    print(f"Entidade: {row['entity']}, Propriedade: {row['property']}, Valor: {row['value']}")

# Consulta SPARQL para verificar as propriedades de proveniência
query_prov = """
PREFIX ex: <http://ivanylsonhg.ppcc.dcc.ice.ufjf.br/enade2022#>
PREFIX prov: <http://www.w3.org/ns/prov#>

SELECT ?entity ?wasGeneratedBy ?wasAssociatedWith ?wasDerivedFrom
WHERE {
    ?entity a ex:Prova .
    ?entity prov:wasGeneratedBy ?wasGeneratedBy .
    ?wasGeneratedBy prov:wasAssociatedWith ?wasAssociatedWith .
    ?entity prov:wasDerivedFrom ?wasDerivedFrom .
}
"""

# Executar a consulta de proveniência
results_prov = g.query(query_prov)

# Exibir resultados
for row in results_prov:
    print(f"Entidade: {row['entity']}, Gerado por: {row['wasGeneratedBy']}, Associado com: {row['wasAssociatedWith']}, Derivado de: {row['wasDerivedFrom']}")
```

Tarefa de classificação para prever o ano de conclusão do ensino médio (ANO_FIM_EM) de estudantes, utilizando como features (variáveis explicativas) o código do curso (CO_CURSO), o ano de ingresso na graduação (ANO_IN_GRAD), e o turno da graduação (CO_TURNO_GRADUACAO).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Carregar dados
file_path = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/microdados2022_arq2.txt'
df = pd.read_csv(file_path, sep=';', encoding='latin1', decimal='.', na_values='""')

# Exemplo de seleção de colunas e limpeza de dados
df = df[['CO_CURSO', 'ANO_FIM_EM', 'ANO_IN_GRAD', 'CO_TURNO_GRADUACAO']]  # Ajuste as colunas conforme necessário

# Tratar valores ausentes
df = df.dropna()

# Convertendo variáveis categóricas para numéricas
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Definir features e target
X = df.drop('ANO_FIM_EM', axis=1)  # Features
y = df['ANO_FIM_EM']  # Target

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Inicializar e treinar o modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

O resultado mostra que a acurácia é baixa (28.6%), e o modelo falhou em prever corretamente várias classes, conforme indicado pelos avisos (UndefinedMetricWarning). A precisão, recall e f1-score são baixos para muitas classes, especialmente aquelas com poucas amostras.

Isso ocorre porque:

Existem muitas classes para a variável ANO_FIM_EM, o que torna o problema desbalanceado.
Algumas classes têm poucos exemplos, dificultando a previsão precisa.

Código faz a atualização de uma ontologia OWL com os resultados de um modelo preditivo (Machine Learning) utilizando a biblioteca rdflib para manipular gráficos RDF.

```python
from rdflib import Graph, URIRef, Literal, Namespace

# Carregar a ontologia existente
g = Graph()
g.parse('/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/enade2022_updated_with_prov.owl', format='xml')

# Definir o namespace
EN = Namespace("http://www.example.org/enade#")

# Mapear índices de X_test para os índices de y_pred
indices = X_test.index.tolist()

# Atualizar a ontologia com os resultados do modelo
for idx, pred_class in zip(indices, y_pred):
    row = X_test.loc[idx]
    student_uri = URIRef(f"http://www.example.org/enade#{row['CO_CURSO']}_{row['ANO_IN_GRAD']}")

    # Adicionar a previsão como uma propriedade
    g.add((student_uri, EN['predictedPerformance'], Literal(pred_class)))

# Salvar a ontologia atualizada
g.serialize(destination='/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/enade2022_updated_with_prov_updated.owl', format='xml')
```

Para classificar estudantes em grupos com base em suas características:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Carregar dados
df = pd.read_csv('/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/microdados2022_arq2.txt', sep=';', encoding='utf-8')

# Verificar colunas disponíveis
print(df.columns)

# Remover linhas com valores faltantes
df = df.dropna(subset=['NU_ANO', 'CO_CURSO', 'ANO_FIM_EM', 'ANO_IN_GRAD', 'CO_TURNO_GRADUACAO'])

# Selecionar características e alvo novamente
X = df[['NU_ANO', 'CO_CURSO', 'ANO_FIM_EM', 'ANO_IN_GRAD']]
y = df['CO_TURNO_GRADUACAO']

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar o modelo
print(classification_report(y_test, y_pred))
```

Resultado do Relatório de Classificação

O resultado do relatório de classificação fornece uma visão detalhada da performance do modelo de RandomForestClassifier em classificar estudantes em grupos com base nas características fornecidas. Aqui está uma explicação do que cada métrica significa e o que você pode inferir dos resultados:

1. Métricas de Avaliação
Precisão (Precision)
Mede a proporção de previsões positivas corretas em relação ao total de previsões positivas feitas pelo modelo.
Exemplo para Categoria 1.0: O modelo acertou 55% das vezes em que previu a categoria 1.0.
Revocação (Recall)
Mede a proporção de verdadeiros positivos em relação ao total de verdadeiros positivos e falsos negativos (ou seja, quantas das instâncias reais da categoria foram corretamente identificadas pelo modelo).
Exemplo para Categoria 1.0: O modelo identificou corretamente 45% das instâncias reais da categoria 1.0.
F1-Score
A média harmônica entre precisão e revocação, fornecendo uma única métrica que combina ambas.
Exemplo para Categoria 1.0: O F1-Score de 0.50 para a categoria 1.0 é uma média entre precisão e revocação.
Suporte (Support)
O número de instâncias reais de cada categoria no conjunto de teste.
Exemplo para Categoria 1.0: Existem 23.443 instâncias reais da categoria 1.0 no conjunto de teste.
Acurácia
A proporção geral de previsões corretas em relação ao total de previsões.
Valor: 83%, indicando que o modelo fez previsões corretas para 83% das instâncias no conjunto de teste.
Macro Average
Média das métricas de precisão, revocação e F1-Score calculadas para cada categoria, sem considerar o suporte (o tamanho das categorias).
Exemplo: Precisão média de 0.71, revocação média de 0.67 e F1-Score médio de 0.69.
Weighted Average
Média das métricas ponderada pelo suporte das categorias, refletindo a distribuição real das categorias no conjunto de teste.
Exemplo: Precisão ponderada de 0.82, revocação ponderada de 0.83 e F1-Score ponderado de 0.83.

2. Interpretação dos Resultados
Bom Desempenho em Categorias Altas
O modelo tem um desempenho forte nas categorias 3.0 e 4.0, com alta precisão e revocação. Isso sugere que as características dos dados para essas categorias são bem diferenciadas e bem representadas no modelo.
Desempenho Inferior em Categorias Baixas
O desempenho é mais baixo nas categorias 1.0 e 2.0. Isso pode indicar que essas categorias são menos distintas ou que o modelo tem dificuldade em distinguir entre elas. Pode ser necessário investigar mais sobre essas categorias, ajustar as características usadas ou aumentar os dados de treinamento.
Acurácia Geral Alta

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Carregar os dados
df = pd.read_csv('/content/drive/MyDrive/Mestrado/TrabalhoESeBD/DADOS/microdados2022_arq2.txt', sep=';', encoding='utf-8')

# Verificar colunas disponíveis
print(df.columns)

# Remover linhas com valores faltantes
df = df.dropna(subset=['NU_ANO', 'CO_CURSO', 'ANO_FIM_EM', 'ANO_IN_GRAD', 'CO_TURNO_GRADUACAO'])

# Preparar os dados
# Use 'CO_TURNO_GRADUACAO' como variável alvo e o restante como características
X = df[['NU_ANO', 'CO_CURSO', 'ANO_FIM_EM', 'ANO_IN_GRAD']]
y = df['CO_TURNO_GRADUACAO']

# Dividir os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar o modelo
print(f'Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
```

Mean Squared Error (MSE) é uma métrica de avaliação que mede a média dos quadrados dos erros (diferença entre os valores previstos e os valores reais). Um MSE menor indica que o modelo está fazendo previsões mais próximas dos valores reais.

O valor do MSE que você obteve é 1.069161974414576. Sem contexto adicional, é difícil interpretar se esse valor é bom ou ruim. Em geral:

Menor MSE: Melhor desempenho do modelo.
Maior MSE: Pior desempenho do modelo.
