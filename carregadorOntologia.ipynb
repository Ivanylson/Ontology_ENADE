Mapear os Dados dos Microdados para a Ontologia

```python
from rdflib import Graph, RDF, RDFS, OWL

# Definir o caminho para a ontologia
ontology_path = '/content/drive/MyDrive/Mestrado/TrabalhoESeBD/enade.owl'

# Criar um grafo e carregar a ontologia
g = Graph()
g.parse(ontology_path, format='xml')

# Verificar o número de tripletas no grafo
print(f"Number of triples in the graph: {len(g)}")

# Verificar as classes definidas na ontologia
print("Classes:")
for s in g.subjects(RDF.type, OWL.Class):
    label = list(g.objects(s, RDFS.label))
    print(s, "->", label if label else "No label found")

# Verificar as propriedades de dados
print("\nPropriedades de Dados:")
for s in g.subjects(RDF.type, OWL.DatatypeProperty):
    label = list(g.objects(s, RDFS.label))
    print(s, "->", label if label else "No label found")

# Verificar as propriedades de objetos
print("\nPropriedades de Objetos:")
for s in g.subjects(RDF.type, OWL.ObjectProperty):
    label = list(g.objects(s, RDFS.label))
    print(s, "->", label if label else "No label found")
```
